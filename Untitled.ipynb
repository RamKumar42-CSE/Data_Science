```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('F:/Data Science/climate.csv')
```

```python
df.head()
```

# Standardising the variables:

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('date',axis=1))
```

```python
scaled_features = scaler.transform(df.drop('date',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(df_feat)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['unnamed 1', 'unnamed 2'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['meanpressure']]], axis = 1)
finalDf
```

```python
finalDf.dropna()
```

# Splitting the data into training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_feat
y=df['date']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

# Predictions and Evaluations

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
